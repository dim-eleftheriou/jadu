```python
import requests
from utils import read_data
```

```python
# Read image data
data = read_data('data/image/qc_results.json')
```

```python
res = requests.post('http://localhost:8080/calculate_risk', json=data[0])
```

```python
res.json()
```

```python
results = []
ind = 0

for i in range(0, len(data)):
    res = requests.post('http://localhost:8080/calculate_risk', json=data[i])
    if res.status_code==200:
        results.append(res.json()["execution_time"])
        ind += 1
    if ind==1000:
        break
```
